```python
import pandas as pd
from datetime import datetime
import numpy as np
import time
```

```python
btcData = pd.read_csv("btcAllData-inclusiveDate.csv")
btcData.head()
```

### bitcoin halving dates

-  Nov. 28, 2012 (ab da 25)
- July 9, 2016 (ab da 12.5)
- May 11, 2020 (ab da 6.25)
- 02.04.2024 irgendwann (ab da 3.125) 

```python
first = "28/11/2012"
second = "09/07/2016"
third = "11/05/2020"
fourth = "02/04/2024"
firstTimestamp =time.mktime(datetime.strptime(first, "%d/%m/%Y").timetuple())
secondTimestamp = time.mktime(datetime.strptime(second, "%d/%m/%Y").timetuple())
thirdTimestamp = time.mktime(datetime.strptime(third, "%d/%m/%Y").timetuple())
fourthTimestamp = time.mktime(datetime.strptime(fourth, "%d/%m/%Y").timetuple())
halvingdates = [firstTimestamp, secondTimestamp, thirdTimestamp, fourthTimestamp]
halvingdates
```

```python
def calculateTimeToNextHalving(timestamp, halvingdates):
    for i in halvingdates:
        if i >= timestamp:
            return int((i - timestamp)/ (24*60*60))
    return None
```

```python
btcData["timeToNextHalving"] = [calculateTimeToNextHalving(x,halvingdates) for x in btcData["time"]]
btcData.head()
```

```python
btcData.to_csv('btcAllData-inclusiveDate-inclusiveHalving.csv',index=False)
```
